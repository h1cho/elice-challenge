# 수열 복원

- 시간 제한: 1초

양의 정수로 이루어진 수열 $*_1,a_2,⋯,a_n$이 있습니다.

$$
1≤a_i≤10^5
$$

이 수열에서 각 원소를 선택하거나 선택하지 않음으로써 총 $2^n$개의 부분 수열을 만들 수 있고, 만들어진 모든 부분 수열의 합인 $2^n$개의 정수가 주어졌을 때, 원래의 수열 $a_1,a_2,⋯,a_n$을 구하는 프로그램을 작성하세요.

---

## **지시사항**

### **입력**

- 첫째 줄에 정수 $n$이 주어집니다.
    
    $$
    1≤n≤15
    $$
    
- 둘째 줄에 이 수열에서 만들 수 있는 모든 부분 수열의 합인 $2^n$개의 정수 $s_{1},s_{2},⋯,s_{2^n}$이 주어집니다.
    
    $$
    0≤s_i≤n×10^5
    $$
    

### **출력**

- 첫째 줄에 원래 수열의 원소를 오름차순으로 출력합니다.

### **입력 예시**

```
3
1 4 7 3 0 6 5 2
```

### **출력 예시**

```
1 2 4
```

````python
n = int(input())
inputs = sorted(list(map(int, input().split())))

# 입력받은 수열 중 0을 제외한 가장 작은 수는 원래 수열의 원소
seq = [inputs[1]]  # 원래 수열
sums = inputs[:2]  # seq의 원소들로 조합한 모든 합
inputs = inputs[2:]

# inputs에 남은 원소가 없을 때까지 반복
while inputs:
    a = inputs[0]  # 다음으로 찾은 원래 수열의 원소
    sums_a = []    # a를 더한 합계 리스트
    for sm in sums:
        new_sm = a + sm
        if new_sm in inputs:
            sums_a.append(inputs.pop(inputs.index(new_sm)))  # new_sm을 inputs에서 제거, sums_a에 추가
    
    # 업데이트 : a와 a가 더해진 합계를 추가
    seq.append(a)
    sums += sums_a

for i in seq:
    print(i, end=" ")
````
